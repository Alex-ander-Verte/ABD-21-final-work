```python
#Фильтруем данные по колонке user_id по значению 405 (user_id самого активного пользователя согласно предыдущему запросу)
df_ratings[df_ratings.user_id == 405]
```

Для построения модели нам нужны признаки. В качестве таковых будем использовать:
+ Год выхода
+ Жанры
+ Общее количество оценок
+ Суммарную оценку


4. Добавьте к датафрейму из задания 3 столбцы:
+ По жанрам. Каждый столбец - это жанр. Единицу записываем, если фильм принадлежит данному жанру и 0 - если нет
+ столбцы с общим количеством оценок от всех пользователей на фильм и суммарной оценкой от всех пользователей


```python
#Создаём новый датафрейм на основе данных из предыдущего запроса
df_user_id_405 =  df_ratings[df_ratings.user_id == 405]
df_user_id_405.head()
```

```python
#Оставляем только нужные колонки идентификатора фильма и рейтинга
df_user_id_405 = df_user_id_405[['item_id', 'rating']]
df_user_id_405.head()
```

```python
#Подключаем библиотеку datetime
import datetime
```

```python
#Выделяем год выхода фильма из даты релиза
df_movies['release_year'] = pd.to_datetime(df_movies['release_date']).dt.strftime('%Y')
df_movies.head()
```

```python
#Создаём датафрейм с годом выхода фильма и жанрами
df_movies_genre = df_movies[["movie_id", "release_year", "unknown", "Action", "Adventure", "Animation", "Children's", "Comedy", "Crime", "Documentary", "Drama", "Fantasy", "Film-Noir", "Horror", "Musical", "Mystery", "Romance", "Sci-Fi", "Thriller", "War", "Western"]]
df_movies_genre.head()
```

```python
#Добавляем вычисляемые колонки "общее кол-во оценок" и "суммарная оценка" в df_ratings
df_ratings['count_rat'] = df_ratings[['item_id', 'rating']].groupby(['item_id'])['rating'].transform('count')
df_ratings['sum_rat'] = df_ratings[['item_id', 'rating']].groupby(['item_id'])['rating'].transform('sum')
df_ratings.head() 
```

```python
#Убираем ненужные колонки для последующего обогащения 
df_rat_merge = df_ratings[['item_id', 'count_rat', 'sum_rat']] 
df_rat_merge.head()
```

```python
#Обогащаем df_user_id_405 данными по году выхода и по жанрам, и создаём итоговый датафрейм
df_user_id_405_final = pd.merge(df_user_id_405, df_movies_genre, how= 'left', left_on='item_id', right_on='movie_id')
df_user_id_405_final.head()
```

```python
#Поскольку item_id и movie_id это одно и тоже, то оставляем только  item_id для дальнейшей работы
df_user_id_405_final = df_user_id_405_final[["item_id", "rating",	"release_year", "unknown", "Action", "Adventure", "Animation", "Children's", "Comedy", "Crime", "Documentary", "Drama", "Fantasy", "Film-Noir", "Horror", "Musical", "Mystery", "Romance", "Sci-Fi", "Thriller", "War", "Western"]]
df_user_id_405_final.head()
```

```python
#Обогащаем df_user_id_405_final данными по оценкам
df_user_id_405_final = pd.merge(df_user_id_405_final, df_rat_merge, how = 'left', on = 'item_id')
df_user_id_405_final.head()
```

#Построение модели
5. Сформируйте X_train, X_test, y_train, y_test

```python
#Подключаем библиотеку sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
#Формируем признаки и целевую переменную
X = df_user_id_405_final[["item_id", "release_year", "unknown", "Action", "Adventure", "Animation", "Children's", "Comedy", "Crime", "Documentary", "Drama", "Fantasy", "Film-Noir", "Horror", "Musical", "Mystery", "Romance", "Sci-Fi", "Thriller", "War", "Western", "count_rat", "sum_rat"]] 
y = df_user_id_405_final[['rating']]
```

```python
#Разделяем данные на выборку для обучения/проверки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

6. Возьмите модель линейной регрессии (или любую другую для задачи регрессии)  и обучите ее на фильмах

```python
#Создаём модель линейной регрессии
lr = LinearRegression()
```

```python
#И обучаем её
lr.fit(X_train, y_train)
```

7. Оцените качество модели на X_test, y_test при помощи метрик для задачи регрессии

```python
#Получаем предсказание
lr.coef_
```

```python
#Оцениваем качество при помощи метода mean_squared_error
mean_squared_error(y_test, lr.predict(X_test))
```

```python
mean_squared_error(y_train, lr.predict(X_train))
```
